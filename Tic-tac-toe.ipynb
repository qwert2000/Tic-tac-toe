<h1 style="text-align:center;">UTILITY FUNCTION </h1>

```python
def utility(state,row,col,pos):
    if win_state(state, COMP,row,col,pos):
        score = +10
    elif win_state(state, HUMAN,row,col,pos):
        score = -10
    else:
        score = 0
    return score
```

<h1 style="text-align:center;">WIN_STATE FUNCTION</h1>

```python
def win_state(state, player,row,col,pos):
    x = player
    if rows_win(state,x,row,col,pos) or cols_win(state,x,row,col,pos) or diag_win(state,x,row,col,pos):
        return True
    else :
        return False
```

<h1 style="text-align:center;">ROWS_WIN FUNCTION</h1>

```python
def rows_win(state,x,row,col,pos):
    for j in range(0,col-pos+1):
        count=0
        for i in range(0,row):
            count=0
            for k in range(0,pos):
                if state[i][k+j] == x:
                    count = count+1
            if count == pos:
                return True
```

<h1 style="text-align:center;">COLS_WIN FUNCTION</h1>

```python
def cols_win(state,x,row,col,pos):
    for j in range(0,row-pos+1):
        count=0
        for i in range(0,col):
            count=0
            for k in range(0,pos):
                if state[k+j][i] == x:
                    count=count+1
            if count == pos:
                return True
```

<h1 style="text-align:center;">DIAG_WIN FUNCTION</h1>

```python
def diag_win(state,x,row,col,pos):
    for k in range(0,row-pos+1):
        count=0
        for i in range(0,pos):
            if state[i+k][i+k]==x:
                count=count+1
        if count==pos:
            return True
    for k in range(0,row-pos+1):
        count=0
        for i in range(0,pos):
            if state[i+k][col-1-i-k] == x:
                count=count+1
        if count==pos:
            return True
    for j in range(1,row-pos+1):
        count=0
        for i in range(0,pos):
            if state[i+j][i] == x:
                count = count+1
        if count == pos:
            return True
    for j in range(1,row-pos+1):
        count =0
        for i in range(0,pos):
            if state[i][i+j] == x:
                count = count+1
        if count == pos:
            return True
    for j in range(1,row-pos+1):
        count=0
        for i in range(0,pos):
            if state[i][col-1-i-j] == x:
                count=count+1
        if count==pos:
            return True
    for j in range(1,row-pos+1):
        count=0
        for i in range(0,pos):
            if state[i+j][col-1-i] == x:
                count=count+1
        if count==pos:
            return True
```

<h1 style="text-align:center;">GAME_OVER FUNCTION</h1>

```python
def game_over(state,row,col,pos):
    return win_state(state, HUMAN,row,col,pos) or win_state(state, COMP,row,col,pos)
```

<h1 style="text-align:center;">EMPTY_POSITIONS FUNCTION</h1>

```python
def empty_positions(state,row,col):
    count=0
    for i in range(0,row):
        for j in range(0,col):
            if state[i][j] == 0:
                count=count+1
    return count
```

<h1 style="text-align:center;">SIMPLE MINIMAX FUNCTION</h1>

```python
def minimax(state, depth, player,row,col,pos):
    if player == COMP:
        maximum = [-1, -1, -infinity]
    else:
        maximum = [-1, -1, +infinity]
    if depth == 0 or game_over(state,row,col,pos):
        score = utility(state,row,col,pos)
        return [-1, -1, score]
    for i in range(0,row):
        for j in range(0,col):
            if state[i][j] == 0:
                x, y = i,j
                state[x][y] = player
                score = minimax(state, depth - 1, -player,row,col,pos)
                state[x][y] = 0
                score[0], score[1] = x, y
                if player == COMP:
                    if score[2] > maximum[2]:
                        maximum = score
                else:
                    if score[2] < maximum[2]:
                        maximum = score
    return maximum
```

<h1 style="text-align:center;">MINIMAX WITH ALPHA BETA PRUNING FUNCTION </h1>

```python
def minimax1(state, depth, player,row,col,pos,alpha,beta):
    if player == COMP:
        maximum = [-1, -1, -infinity]
    else:
        maximum = [-1, -1, +infinity]
    if depth == 0 or game_over(state,row,col,pos):
        score = utility(state,row,col,pos)
        return [-1, -1, score]
    for i in range(0,row):
        for j in range(0,col):
            if state[i][j] == 0:
                x, y = i,j
                state[x][y] = player
                score = minimax1(state, depth - 1, -player,row,col,pos,alpha,beta)
                state[x][y] = 0
                score[0], score[1] = x, y
                if player == COMP:
                    if score[2] > maximum[2]:
                        maximum = score
                    if maximum[2] > alpha:
                        alpha = maximum[2]
                    if alpha >= beta:
                        break
                else:
                    if score[2] < maximum[2]:
                        maximum = score
                    if maximum[2] < beta:
                        beta = maximum[2]
                    if alpha >= beta:
                        break
    return maximum
```

```python
def minimax2(state, depth, player,row,col,pos,depth1):
    if player == COMP:
        maximum = [-1, -1, -infinity]
    else:
        maximum = [-1, -1, +infinity]
    if depth == depth1 or game_over(state,row,col,pos):
        score = evaluvate(state,row,col,pos)
        return [-1, -1, score]
    for i in range(0,row):
        for j in range(0,col):
            if state[i][j] == 0:
                x, y = i,j
                state[x][y] = player
                score = minimax2(state, depth - 1, -player,row,col,pos,depth1)
                state[x][y] = 0
                score[0], score[1] = x, y
                if player == COMP:
                    if score[2] > maximum[2]:
                        maximum = score
                else:
                    if score[2] < maximum[2]:
                        maximum = score
    return maximum
```

```python
def minimax3(state, depth, player,row,col,pos,depth1,alpha,beta):
    if player == COMP:
        maximum = [-1, -1, -infinity]
    else:
        maximum = [-1, -1, +infinity]
    if depth == depth1 or game_over(state,row,col,pos):
        score = evaluvate(state,row,col,pos)
        return [-1, -1, score]
    for i in range(0,row):
        for j in range(0,col):
            if state[i][j] == 0:
                x, y = i,j
                state[x][y] = player
                score = minimax3(state, depth - 1, -player,row,col,pos,depth1,alpha,beta)
                state[x][y] = 0
                score[0], score[1] = x, y
                if player == COMP:
                    if score[2] > maximum[2]:
                        maximum = score
                    if maximum[2] > alpha:
                        alpha = maximum[2]
                    if alpha >= beta:
                        break
                else:
                    if score[2] < maximum[2]:
                        maximum = score
                    if maximum[2] < beta:
                        beta = maximum[2]
                    if alpha >= beta:
                        break
    return maximum
```

<h1 style="text-align:center;">MINIMAX WITH HEURISTIC FUNCTION</h1>

```python
def minimax4(state, depth, player,row,col,pos):
    if player == COMP:
        maximum = [-1, -1, -infinity]
    else:
        maximum = [-1, -1, +infinity]
    if depth == 0 or game_over(state,row,col,pos):
        score = evaluvate(state,row,col,pos)
        return [-1, -1, score]
    x = [-1,-1,-infinity]
    for i in range(0,row):
        for j in range(0,col):
            if state[i][j] == 0:
                state[i][j] = player
                y = evaluvate(state,row,col,pos)
                state[i][j] = 0
                if y > x[2]:
                    x[2] = y
                    x[0] = i
                    x[1] = j
    a = x[0]
    b =x[1]
    state[a][b] = player
    score = minimax4(state, depth - 1, -player,row,col,pos)
    state[a][b] = 0
    score[0], score[1] = a, b
    if player == COMP:
        if score[2] > maximum[2]:
            maximum = score
        else:
            if score[2] < maximum[2]:
                maximum = score
    return maximum
```
